Métodos de string: 
As strings oferecem métodos que executam várias operações úteis. Métodos são comandos colocados ao final da string para realizar determinadas ações e modificações.
Dentre os principais métodos aplicáveis a strings e suas funcionalidades podemos citar:

exemplo.upper(): retorna a string 'exemplo' em letras maiúsculas;

exemplo.lower(): retorna a string 'exemplo' toda em letras minúsculas;

exemplo.strip(): retorna o 'exemplo' retirando possíveis espaços em branco no início e no fim da string;

exemplo.startswith('x'): retorna se o 'exemplo' começa com a string 'x';

exemplo.endswith('z'): retorna se o 'exemplo' termina com a string 'z';

exemplo.replace('e','i'): retorna a string com todas as letras 'e' em 'Dev' substituídas por 'i'.

exemplo.find('xyz'): procura a string 'xyz' dentro de 'exemplo' e retorna o primeiro índice onde 'xyz' começa ou retorna -1 se nada for encontrado;

```python
#upper
exemplo = "xadrez"
exemplo.upper()
```

```python
#replace
exemplo2 = "Programar é muito chato"
exemplo2.replace('chato' , 'bom')
```

Input e Output
Veremos agora como utilizar informações externas no código!
Para perguntar informações para o usuário, utiliza-se a função "input()"

```python
# A string colocada dentro da função será exibido como mensagem para o usuário
input("Como você está?")
```

```python
# O Output, é exibir mensagens para o usuário
print("Continue estudando")
```

Podemos atribuir inputs à variáveis para trabalhar com os dados

```python
# \n  (pula uma linha)
idade = input("Quantos anos você tem? \n")
```

```python
# Duas maneiras de colocar variáveis com strings em prints:
print("Você possui", idade, "anos, felicidades")
```

```python
print(f'Você possui {idade} anos, que legal')
```

f_strings
Esta string neste último print que vimos é chamado de f_string. Ao colocar um f na frente, podemos adicionar variáveis, operações, fórmulas e métodos dentro de {} nas strings!

```python
a = 5
b = 2
bebida = "chocolates"
f_string = f"Tomei incríveis {a*b} xícaras de {bebida.upper() + a*'!'}"
print(f_string)
```
